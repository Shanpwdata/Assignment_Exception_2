Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

#Ans_1.
Python provides a built-in Exception class, which serves as the base class for all built-in and user-defined exceptions thats benifically to his properties of class like (Inheritance).

```python
#Example:
import logging
logging.basicConfig(filename = "error.log", level = logging.ERROR)
class CustomError(Exception):
        def __init__(self,msg):
            self.msg = msg
try:
    num = int(input("Enter a number:"))
    if num < 0:
        raise CustomError("Negative numbers are not allowed.")
except CustomError as e:
    logging.error("This is a Custom Error:{}" .format(e))
except Exception as e:    # inherit the error from Exception class
    logging.error("This is a General Exception:{}".format(e))
```

Q2. Write a python program to print Python Exception Hierarchy.

```python
#Ans_2.
# import inspect module
import inspect

# our treeClass function
def treeClass(cls, ind = 0):
	
	# print name of the class
	print ('-' * ind, cls.__name__)
	
	# iterating through subclasses
	for i in cls.__subclasses__():
		treeClass(i, ind + 3)

print("Hierarchy for Built-in exceptions is : ")

# inspect.getmro() Return a tuple
# of class cls’s base classes.

# building a tree hierarchy
inspect.getclasstree(inspect.getmro(Exception))

# function call
treeClass(Exception)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

#Ans_5.
In Python, ImportError is an exception that is raised when an import statement fails to locate and load the module or package that is being imported

```python
#Example:- ImportError
try :
    import shantam
except ImportError as e:
    print(e)
```

ModuleNotFoundError :-- In Python 3.6 and later versions, when an import statement fails to locate the specified module, Python raises a ModuleNotFoundError instead of a generic ImportError.

```python
#Example:-
try:
    import non_exist_module
except ModuleNotFoundError as e:
    print("Module Not Found Error: ", e)
```
